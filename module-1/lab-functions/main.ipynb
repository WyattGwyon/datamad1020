## 2. Now write a function that returns the largest element on a list

```python
def greatest(arr):
    largest = arr[0]
    for num in arr:
        if num > largest:
            largest = num
    return largest
```

```python
# My test
a = [-1,-2,-3,-4,-5,-6,-7,-8]
print(greatest(a))
```

## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
    if n > 0:
        fact = n
        n_minus_1 = n-1
    while n_minus_1 > 0:    
            fact *= n_minus_1
            n_minus_1 -= 1
            
    return fact
           
        
```

```python
# My test
print(factorial(5))
```

```python
# This will test your function 
test_factorial(factorial)
```

## 7. Write a function that takes a list and returns a list of the unique values.

`NOTE: You cannot use set. ü§î`

```python
def unique(arr):
    b = []
    for element in arr:
        if element not in b:
            b += [element]
    return b
```

## 9. Write a function that calculates the standard deviation of a list.
`NOTE: Do not use any libraries or already built functions. üòâ`

```python
def st_dev(arr):
# Total and count of arr
    tot = 0
    count = 0
    for num in arr:
        tot += num
        count += 1
    print("tot", tot)
    print("count", count)

# Mean of arr    
    mean = tot/count
    print("mean", mean)

# List of deviations of arr    
    dev = [(num - mean)**2 for num in arr]
    print(dev)

# Sum of deviations    
    dev_sum = 0
    for d in dev:
        dev_sum += d
    print("dev_sum",dev_sum)
    
# Variance of arr        
    var = dev_sum/count
    print("var",var)
    
# Standard Deviation of arr    
    std_d = var **.5
    print("std_d",std_d)
    
    return std_d
```

```python
# My test
a = [-1,-2,-3,-4,-5,-6,-7,-8, 9]
st_dev(a)
```

```python
# This will test your function 
test_stdev(st_dev)
```

## 10. Write a function to check if a string is a pangram, i.e.: if it contains all the letters of the alphabet at least once.

```python
def pangram(string):
    alpha = []
    for letter in string:
        if letter not in alpha:
            alpha += [letter]
    count = 0
    for l in alpha:
        count += 1
        
    return bool(count >= 26)
```
